```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File path to write the transformed Apple Store and Google Play Store apps dataset in .csv format

Output_dir_path = "./output/"
directory = os.path.dirname(Output_dir_path)
```

```python
# Import the Google Play Store data

gps_data = pd.read_csv(os.path.join('Resources','Google_data_ext.csv'), index_col=0)
gps_data.head()
```

```python
# Display pre_transformed 'Category' data

gps1DF = gps_data['Category']
gps1DF = gps1DF.drop_duplicates()
print(gps1DF)
```

```python
# Transform the Google play store 'Category' data to align with Apple store 'prime-genere' as per the following mapping

gps_data['Category'].replace('AUTO_AND_VEHICLES', 'Utilities',inplace=True)
gps_data['Category'].replace('BEAUTY', 'LIFESTYLE',inplace=True)
gps_data['Category'].replace('COMICS', 'BOOKS_AND_REFERENCE',inplace=True)
gps_data['Category'].replace('COMMUNICATION', 'Music',inplace=True)
gps_data['Category'].replace('DATING', 'SOCIAL',inplace=True)
gps_data['Category'].replace('EVENTS', 'Catalogs',inplace=True)
gps_data['Category'].replace('FAMILY', 'LIFESTYLE',inplace=True)
gps_data['Category'].replace('HOUSE_AND_HOME', 'Catalogs',inplace=True)
gps_data['Category'].replace('LIBRARIES_AND_DEMO', 'Reference',inplace=True)
gps_data['Category'].replace('HPARENTING', 'LIFESTYLE',inplace=True)
gps_data['Category'].replace('PHOTOGRAPHY', 'Photo & Video',inplace=True)
gps_data['Category'].replace('TOOLS', 'Catalogs',inplace=True)
gps_data['Category'].replace('VIDEO_PLAYERS', 'Photo & Video',inplace=True)
```

```python
# Display Transformed 'Category' data

gps2DF = gps_data['Category']
gps2DF = gps2DF.drop_duplicates()
print(gps2DF)
```

```python
# Display pre_transformed 'Content_Rating' data

gps1DF = gps_data['Content_Rating']
gps1DF = gps1DF.drop_duplicates()
print(gps1DF)
```

```python
# Transform the Google play store 'Content Rating' data to align with Apple store 'cont-rating'

gps_data['Content_Rating'].replace('Teen', '12+', inplace=True)
gps_data['Content_Rating'].replace('Everyone 10+', '9+', inplace=True)
gps_data['Content_Rating'].replace('Adults only 18+', '17+', inplace=True)
gps_data['Content_Rating'].replace('Unrated', '4+', inplace=True)
gps_data['Content_Rating'].replace('Everyone', '17+', inplace=True)
gps_data['Content_Rating'].replace('Mature 17+', '17+', inplace=True)
gps_data['Content_Rating'].fillna('17+', inplace=True)
gps_data['User_Rating'].fillna(0.0, inplace=True)
```

```python
# Display Transformed 'Content_Rating' data

gps2DF = gps_data['Content_Rating']
gps2DF = gps2DF.drop_duplicates()
print(gps2DF)
```

```python
# Create the 'Output' directory in the PWD, if it is not already existing

if not os.path.exists(directory):
   os.makedirs(directory)

gpstrsFile =  Output_dir_path + "Google_data_ext_trf.csv"

# Remove the Google playstore transformed file from the 'Output' directory, if it is already existing

if os.path.isfile(gpstrsFile):
  os.remove(gpstrsFile) 

# Save transformed .csv file with the file name and path provided

gps_data.to_csv(gpstrsFile)
```

```python
gps_data['App_Store'] = 'Google'
gps_data1 = gps_data[['App_Store','App_Name','App_Size','currency','App_Price','Type','Review_count','User_Rating','Content_Rating','Category']]
gps_data1.head()
```

```python
gps_data2=pd.pivot_table(gps_data1, index=['App_Store','Category','Type', 'Content_Rating'],values=['App_Price','User_Rating','Review_count'],aggfunc = np.average)
pd.options.display.float_format = '{:.2f}'.format
gps_data2 = gps_data2.rename(
   columns={"App_Price": "Avg_App_Price",
            "Review_count": "Avg_review_count",
            "User_Rating": "Avg_User_Rating"
           })
gps_data2.head(10)
```

```python
# Create the 'Output' directory in the PWD, if it is not already existing

if not os.path.exists(directory):
   os.makedirs(directory)

gpssumFile =  Output_dir_path + "gpsdata_sum.csv"

# Remove the Google playstore summary file from the 'Output' directory, if it is already existing
if os.path.isfile(gpssumFile):
  os.remove(gpssumFile) 

gps_data2.to_csv(gpssumFile)
```

```python
# Import the apple store data

as_data = pd.read_csv(os.path.join('Resources','Apple_data_ext.csv'), index_col=0)
as_data.head()
```

```python
# Transform / Update the Apple store 'Type' data based on App-price = 0 or not

def conv_type (row):
   if row['App_Price'] == 0 :
      return 'Free'
   else:
      return 'Paid'
```

```python
as_data['Type'] = as_data.apply (lambda row: conv_type(row), axis=1)
as_data.head()
```

```python
# Create the 'Output' directory in the PWD, if it is not already existing

if not os.path.exists(directory):
   os.makedirs(directory)

astrsFile =  Output_dir_path + "Apple_data_ext_trf.csv"

# Remove the Apple store transformed file from the 'Output' directory, if it is already existing

if os.path.isfile(astrsFile):
  os.remove(astrsFile) 

# Save transformed .csv file with the file name and path provided

as_data.to_csv(astrsFile)
```

```python
as_data['App_Store'] = 'Apple'
as_data1 = as_data[['App_Store','App_Name','App_Size','currency','App_Price','Type','Review_count','User_Rating','Content_Rating','Category']]
as_data1.head()
```

```python
as_data2 = pd.pivot_table(as_data1, index=['App_Store', 'Category','Type', 'Content_Rating'],values=['App_Price','User_Rating','Review_count'],aggfunc = np.average)
pd.options.display.float_format = '{:.2f}'.format
as_data2 = as_data2.rename(
   columns={"App_Price": "Avg_App_Price",
            "Review_count": "Avg_Review_Count",
            "User_Rating": "Avg_User_Rating"
           })
as_data2.head()
```

```python
# Create the 'Output' directory in the PWD, if it is not already existing

if not os.path.exists(directory):
   os.makedirs(directory)

assumFile =  Output_dir_path + "asdata_sum.csv"

# Remove the Apple store summary file from the 'Output' directory, if it is already existing

if os.path.isfile(assumFile):
  os.remove(assumFile) 

as_data2.to_csv(assumFile)
```
